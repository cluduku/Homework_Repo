```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("cumulative.csv")
df = df.drop(columns=["rowid", "kepid", "kepoi_name", "kepler_name", "koi_pdisposition", "koi_score", "koi_tce_delivname"])
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
#Get training features
X =  df.values[:,1:]
#Get labels
Y = df.values[:,0]
```

```python
#Encode labels
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
label_encoder.fit(Y)
encoded_y = label_encoder.transform(Y)
```

```python
# Use train_test_split to create training and testing data
from sklearn.model_selection import train_test_split

deep_X_train, deep_X_test, deep_y_train, deep_y_test = train_test_split(X, encoded_y, random_state=1)
```

```python
# Create a MinMaxScaler model and fit it to the training data
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(deep_X_train)
```

```python
# Transform the training and testing data using the X_scaler
X_train_scaled = X_scaler.transform(deep_X_train)
X_test_scaled = X_scaler.transform(deep_X_test)
```

## Model 1: Deep learning

```python
from tensorflow.keras.utils import to_categorical
# One-hot encoding
y_train_categorical = to_categorical(deep_y_train)
y_test_categorical = to_categorical(deep_y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.regularizers import l2

model = Sequential()
model.add(Dense(units=50, activation='relu', input_dim=40,kernel_regularizer=l2(0.001), bias_regularizer=l2(0.001)))
model.add(Dense(units=50, activation='relu',kernel_regularizer=l2(0.001), bias_regularizer=l2(0.001)))
model.add(Dense(units=3, activation='softmax',kernel_regularizer=l2(0.001)))
```

```python
# Compile the model
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
# Fit the model to the training data
model.fit(X_train_scaled,y_train_categorical,epochs=500,shuffle=True,verbose=2)
```

```python
#Evaluate deep learning model
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

## Model 2:  SVM with grid search


```python
#Import dependencies
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import style
style.use("ggplot")
```

```python
# Create the SVC Model
from sklearn.svm import SVC 
model2 = SVC(kernel='linear')
```

```python
model2.fit(X_train_scaled, deep_y_train)
```

```python
print(f"Training Data Score: {model2.score(X_train_scaled, deep_y_train)}")
print(f"Testing Data Score: {model2.score(X_test_scaled, deep_y_test)}")
```

## Hyperparameter tuning

```python
# Create the GridSearch estimator along with a parameter object containing the values to adjust
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5, 10, 50],
              'gamma': [0.0001, 0.0005, 0.001, 0.005]}
grid = GridSearchCV(model2, param_grid, verbose=3)
```

```python
grid.fit(X_train_scaled,np.array(deep_y_train))
```

```python
# List the best parameters for this dataset
print(grid.best_params_)
```
